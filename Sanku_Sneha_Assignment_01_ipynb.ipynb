# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
import math

# Function to calculate hemisphere volume
def hemisphere_volume(r):
    return (2/3) * math.pi * (r**3)

print("r\tV")
for r in range(2, 21, 2):  # Loop through even numbers from 2 to 20
    print(f"{r}\t{hemisphere_volume(r):.2f}")
```

Sentence1: a:2, e:2, i:0, etc.

```python
import string

def get_vowel_count(sentence):
    vowels = 'aeiou'
    vowel_count = {vowel: sentence.lower().count(vowel) for vowel in vowels}
    return vowel_count

def remove_punctuation(sentence):
    return sentence.translate(str.maketrans('', '', string.punctuation))

# (a) Prompt the user to input two sentences
sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

# (b) Transform each sentence into a list of words
words_sentence1 = sentence1.split()
words_sentence2 = sentence2.split()

# Output the list of words from both sentences
print("\nSentence 1 as list of words:", words_sentence1)
print("Sentence 2 as list of words:", words_sentence2)

# (c) Compare the two sentences and find words that occur in both
common_words = list(set(words_sentence1) & set(words_sentence2))
print("\nCommon words in both sentences:", common_words)

# (d) Remove any punctuation and print the sanitized sentences
sanitized_sentence1 = remove_punctuation(sentence1)
sanitized_sentence2 = remove_punctuation(sentence2)

print("\nSanitized Sentence 1:", sanitized_sentence1)
print("Sanitized Sentence 2:", sanitized_sentence2)

# (e) Count the number of vowels in each sentence
vowel_count_sentence1 = get_vowel_count(sanitized_sentence1)
vowel_count_sentence2 = get_vowel_count(sanitized_sentence2)

# Print out the vowel counts
print("\nVowel counts in Sentence 1:", vowel_count_sentence1)
print("Vowel counts in Sentence 2:", vowel_count_sentence2)
```

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

```python
import math

def evaluate_expression(expression):
    try:
        # Evaluate the expression safely
        result = eval(expression)
        return result
    except ZeroDivisionError:
        return "Error: Division by zero"
    except Exception as e:
        return f"Error: Invalid input. {str(e)}"

def calculator():
    print("Simple Command-Line Calculator")
    print("Supported operations: +, -, *, /, ** (exponentiation), % (modulus), sqrt (square root), abs (absolute value), // (floor division), log (logarithm), parentheses ()")
    print("Type 'exit' to quit the calculator.")

    while True:
        expression = input("Enter an expression: ").strip()

        if expression.lower() == 'exit':
            print("Exiting the calculator. Goodbye!")
            break

        # Replace sqrt with math.sqrt and log with math.log (logarithm)
        expression = expression.replace("sqrt", "math.sqrt")
        expression = expression.replace("log", "math.log")
        expression = expression.replace("abs", "math.fabs")  # Absolute value
        expression = expression.replace("//", "/")            # Use / for floor division

        result = evaluate_expression(expression)
        print(f"Result: {result}")

# Run the calculator
calculator()
```

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

To answer this question, you will need to download the file [Assignment1_denton_housing.csv](https://raw.githubusercontent.com/unt-iialab/info5731-spring2022/main/assignments/Assignment1_denton_housing.csv) included with this assignment. The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (5 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import pandas as pd
from scipy.stats import hmean

# Load the CSV file
def load_data(file_path):
    return pd.read_csv(file_path)

def calculate_occupied_difference(df):
    occupied_df = df[df['title_field'] == 'Occupied Housing Units'].sort_values(by='year', ascending=False)
    occupied_df['difference'] = occupied_df['value'].diff(periods=-1)
    occupied_diff_sorted = occupied_df[['year', 'difference']].dropna().sort_values(by='difference', ascending=True)
    return occupied_diff_sorted

def calculate_vacant_occupied_percentages(df):
    housing_pivot = df.pivot(index='year', columns='title_field', values='value')
    housing_pivot['Vacant Housing Units %'] = (housing_pivot['Vacant Housing Units'] / housing_pivot['Total Housing Units']) * 100
    housing_pivot['Occupied Housing Units %'] = (housing_pivot['Occupied Housing Units'] / housing_pivot['Total Housing Units']) * 100
    return housing_pivot[['Vacant Housing Units %', 'Occupied Housing Units %']]

def find_max_vacant_occupied(df):
    housing_pivot = df.pivot(index='year', columns='title_field', values='value')
    max_vacant_year = housing_pivot['Vacant Housing Units'].idxmax()
    max_vacant_value = housing_pivot['Vacant Housing Units'].max()
    max_occupied_year = housing_pivot['Occupied Housing Units'].idxmax()
    max_occupied_value = housing_pivot['Occupied Housing Units'].max()
    return {
        "Vacant Housing Units": {"Year": max_vacant_year, "Value": max_vacant_value},
        "Occupied Housing Units": {"Year": max_occupied_year, "Value": max_occupied_value},
    }

def calculate_harmonic_mean(df):
    total_units = df[df['title_field'] == 'Total Housing Units']['value'].values
    return hmean(total_units)

# Example usage
if __name__ == "__main__":
    file_path = "Assignment1_denton_housing (1).csv"  # Change this to your actual file path
    df = load_data(file_path)

    print("(a) Year-over-Year Differences in Occupied Housing Units:")
    print(calculate_occupied_difference(df))

    print("\n(b) Percentage of Vacant and Occupied Housing Units:")
    print(calculate_vacant_occupied_percentages(df))

    print("\n(c) Years with Maximum Vacant and Occupied Housing Units:")
    print(find_max_vacant_occupied(df))

    print("\n(d) Harmonic Mean of Total Housing Units:")
    print(calculate_harmonic_mean(df))
```

# Inventory Management System

This notebook contains the Python program for an advanced inventory management system. The system allows users to add, delete, update, and search for products. It also includes features for automated restocking and generating inventory reports. The program handles various edge cases and uses a CSV file for data storage.

Below are the prompts and revised prompts used to develop the program, followed by the Python implementation.

**Prompt**

Create a sophisticated Python application for a product management company to handle inventory records.
Along with search capabilities, the system should have features for adding, removing, and updating products.
Automate restocking and store data in CSV files. Deal with faulty CSV files and edge cases.

 **Revised Prompt 2**:

Add detailed comments to the code explaining how each edge case is handled. Include edge cases like:

4. Searching for a product that does not exist.

5. Handling corrupted CSV files (e.g., invalid headers or data types).

6. Attempting to add a product with missing fields.

```python
# Inventory Management System - Python Implementation

import csv
import os

# Define the CSV file path
INVENTORY_FILE = "inventory.csv"

# Define the minimum stock level for restocking
MIN_STOCK_LEVEL = 5

# Define the structure of the inventory CSV file
CSV_HEADER = ["ProductID", "Name", "Category", "Price", "Quantity"]

# Function to load inventory from CSV file
def load_inventory():
    """
    Load inventory data from the CSV file.
    Handles file not found and corrupted CSV files.
    """
    inventory = []
    try:
        with open(INVENTORY_FILE, mode="r", newline="") as file:
            reader = csv.DictReader(file)
            # Edge Case 5: Corrupted CSV file (invalid headers)
            if reader.fieldnames != CSV_HEADER:
                raise ValueError("Corrupted CSV file: Invalid headers.")
            for row in reader:
                # Edge Case 5: Corrupted CSV file (invalid data types)
                try:
                    row["Price"] = float(row["Price"])
                    row["Quantity"] = int(row["Quantity"])
                except ValueError:
                    raise ValueError("Corrupted CSV file: Invalid data types.")
                inventory.append(row)
    except FileNotFoundError:
        # If the file doesn't exist, create it with headers
        with open(INVENTORY_FILE, mode="w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=CSV_HEADER)
            writer.writeheader()
    return inventory

# Function to save inventory to CSV file
def save_inventory(inventory):
    """
    Save inventory data to the CSV file.
    """
    with open(INVENTORY_FILE, mode="w", newline="") as file:
        writer = csv.DictWriter(file, fieldnames=CSV_HEADER)
        writer.writeheader()
        writer.writerows(inventory)

# Function to add a new product
def add_product(inventory, product_id, name, category, price, quantity):
    """
    Add a new product to the inventory.
    Handles duplicate product IDs and missing fields.
    """
    # Edge Case 1: Adding a product that already exists
    for product in inventory:
        if product["ProductID"] == product_id:
            raise ValueError("Product ID already exists.")

    # Edge Case 6: Attempting to add a product with missing fields
    if not all([product_id, name, category, price, quantity]):
        raise ValueError("Missing fields. All fields are required.")

    # Add the new product
    inventory.append({
        "ProductID": product_id,
        "Name": name,
        "Category": category,
        "Price": float(price),
        "Quantity": int(quantity)
    })
    save_inventory(inventory)
    print(f"Product '{name}' added successfully.")

# Function to delete a product
def delete_product(inventory, product_id):
    """
    Delete a product from the inventory.
    Handles product not found.
    """
    # Edge Case 2: Deleting a product that does not exist
    for product in inventory:
        if product["ProductID"] == product_id:
            inventory.remove(product)
            save_inventory(inventory)
            print(f"Product '{product['Name']}' deleted successfully.")
            return
    raise ValueError("Product not found.")

# Function to update a product's quantity
def update_product_quantity(inventory, product_id, quantity):
    """
    Update the quantity of a product.
    Handles negative quantity and product not found.
    """
    # Edge Case 3: Updating the quantity to a negative value
    if quantity < 0:
        raise ValueError("Quantity cannot be negative.")

    # Edge Case 9: Attempting to update a product that does not exist
    for product in inventory:
        if product["ProductID"] == product_id:
            product["Quantity"] = quantity
            save_inventory(inventory)
            print(f"Quantity updated for product '{product['Name']}'.")
            return
    raise ValueError("Product not found.")

# Function to search for a product
def search_product(inventory, product_id):
    """
    Search for a product by ID.
    Handles product not found.
    """
    # Edge Case 4: Searching for a product that does not exist
    for product in inventory:
        if product["ProductID"] == product_id:
            return product
    raise ValueError("Product not found.")

# Function to list all products
def list_products(inventory):
    """
    List all products in the inventory.
    """
    if not inventory:
        print("Inventory is empty.")
        return

    print("\n--- List of Products ---")
    for product in inventory:
        print(f"Product ID: {product['ProductID']}, Name: {product['Name']}, Category: {product['Category']}, Price: ${product['Price']}, Quantity: {product['Quantity']}")

# Function to generate a report
def generate_report(inventory):
    """
    Generate a report showing item values and category breakdowns.
    Handles empty inventory.
    """
    # Edge Case 8: Generating a report when the inventory is empty
    if not inventory:
        raise ValueError("Inventory is empty.")

    total_value = 0
    category_breakdown = {}

    for product in inventory:
        total_value += product["Price"] * product["Quantity"]
        category = product["Category"]
        if category in category_breakdown:
            category_breakdown[category] += product["Price"] * product["Quantity"]
        else:
            category_breakdown[category] = product["Price"] * product["Quantity"]

    print("\n--- Inventory Report ---")
    print(f"Total Inventory Value: ${total_value:.2f}")
    print("Category Breakdown:")
    for category, value in category_breakdown.items():
        print(f"{category}: ${value:.2f}")

# Function to automate restocking
def automate_restocking(inventory):
    """
    Automate restocking for products below the minimum stock level.
    """
    # Edge Case 7: Restocking a product that has reached the minimum stock level
    for product in inventory:
        if product["Quantity"] < MIN_STOCK_LEVEL:
            product["Quantity"] += MIN_STOCK_LEVEL
            print(f"Restocked product '{product['Name']}' to {product['Quantity']} units.")
    save_inventory(inventory)

# Function to initialize the inventory with 50 products
def initialize_inventory():
    """
    Initialize the inventory with 50 sample products.
    """
    inventory = []
    categories = ["Electronics", "Clothing", "Home Appliances", "Books", "Toys"]
    for i in range(1, 51):
        product = {
            "ProductID": f"P{i:03}",  # Format product ID as P001, P002, etc.
            "Name": f"Product {i}",
            "Category": categories[i % len(categories)],  # Cycle through categories
            "Price": round(10 + (i * 5), 2),  # Price increases with product ID
            "Quantity": 10 + (i % 10)  # Quantity varies between 10 and 19
        }
        inventory.append(product)
    save_inventory(inventory)
    print("Inventory initialized with 50 products.")

# Main function to manage the inventory system
def main():
    # Initialize inventory with 50 products if the file is empty
    if not os.path.exists(INVENTORY_FILE) or os.path.getsize(INVENTORY_FILE) == 0:
        initialize_inventory()

    inventory = load_inventory()

    while True:
        print("\n--- Inventory Management System ---")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product Quantity")
        print("4. Search Product")
        print("5. Generate Report")
        print("6. Automate Restocking")
        print("7. List All Products")
        print("8. Exit")

        choice = input("Enter your choice: ")

        try:
            if choice == "1":
                product_id = input("Enter Product ID: ")
                name = input("Enter Product Name: ")
                category = input("Enter Product Category: ")
                price = input("Enter Product Price: ")
                quantity = input("Enter Product Quantity: ")
                add_product(inventory, product_id, name, category, price, quantity)

            elif choice == "2":
                product_id = input("Enter Product ID to delete: ")
                delete_product(inventory, product_id)

            elif choice == "3":
                product_id = input("Enter Product ID to update: ")
                quantity = int(input("Enter new quantity: "))
                update_product_quantity(inventory, product_id, quantity)

            elif choice == "4":
                product_id = input("Enter Product ID to search: ")
                product = search_product(inventory, product_id)
                print(f"Product Found: {product}")

            elif choice == "5":
                generate_report(inventory)

            elif choice == "6":
                automate_restocking(inventory)

            elif choice == "7":
                list_products(inventory)

            elif choice == "8":
                print("Exiting the system. Goodbye!")
                break

            else:
                print("Invalid choice. Please try again.")

        except Exception as e:
            print(f"Error: {e}")

# Run the program
if __name__ == "__main__":
    main()
```

```python
# Function to add products in bulk
def add_products_in_bulk(inventory, products):
    """
    Add multiple products to the inventory in bulk.
    Handles duplicate product IDs and missing fields.
    """
    for product in products:
        product_id = product.get("ProductID")
        name = product.get("Name")
        category = product.get("Category")
        price = product.get("Price")
        quantity = product.get("Quantity")

        # Edge Case 1: Adding a product that already exists
        for existing_product in inventory:
            if existing_product["ProductID"] == product_id:
                raise ValueError(f"Product ID {product_id} already exists.")

        # Edge Case 6: Attempting to add a product with missing fields
        if not all([product_id, name, category, price, quantity]):
            raise ValueError("Missing fields. All fields are required.")

        # Add the new product
        inventory.append({
            "ProductID": product_id,
            "Name": name,
            "Category": category,
            "Price": float(price),
            "Quantity": int(quantity)
        })

    save_inventory(inventory)
    print(f"{len(products)} products added successfully.")

# Function to delete products in bulk
def delete_products_in_bulk(inventory, product_ids):
    """
    Delete multiple products from the inventory in bulk.
    Handles product not found.
    """
    deleted_count = 0
    for product_id in product_ids:
        found = False
        for product in inventory:
            if product["ProductID"] == product_id:
                inventory.remove(product)
                deleted_count += 1
                found = True
                break
        if not found:
            raise ValueError(f"Product ID {product_id} not found.")

    save_inventory(inventory)
    print(f"{deleted_count} products deleted successfully.")

# Main function to manage the inventory system
def main():
    # Initialize inventory with 50 products if the file is empty
    if not os.path.exists(INVENTORY_FILE) or os.path.getsize(INVENTORY_FILE) == 0:
        initialize_inventory()

    inventory = load_inventory()

    while True:
        print("\n--- Inventory Management System ---")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product Quantity")
        print("4. Search Product")
        print("5. Generate Report")
        print("6. Automate Restocking")
        print("7. List All Products")
        print("8. Add Products in Bulk")
        print("9. Delete Products in Bulk")
        print("10. Exit")

        choice = input("Enter your choice: ")

        try:
            if choice == "1":
                product_id = input("Enter Product ID: ")
                name = input("Enter Product Name: ")
                category = input("Enter Product Category: ")
                price = input("Enter Product Price: ")
                quantity = input("Enter Product Quantity: ")
                add_product(inventory, product_id, name, category, price, quantity)

            elif choice == "2":
                product_id = input("Enter Product ID to delete: ")
                delete_product(inventory, product_id)

            elif choice == "3":
                product_id = input("Enter Product ID to update: ")
                quantity = int(input("Enter new quantity: "))
                update_product_quantity(inventory, product_id, quantity)

            elif choice == "4":
                product_id = input("Enter Product ID to search: ")
                product = search_product(inventory, product_id)
                print(f"Product Found: {product}")

            elif choice == "5":
                generate_report(inventory)

            elif choice == "6":
                automate_restocking(inventory)

            elif choice == "7":
                list_products(inventory)

            elif choice == "8":
                products = []
                while True:
                    product_id = input("Enter Product ID (or 'done' to finish): ")
                    if product_id.lower() == 'done':
                        break
                    name = input("Enter Product Name: ")
                    category = input("Enter Product Category: ")
                    price = input("Enter Product Price: ")
                    quantity = input("Enter Product Quantity: ")
                    products.append({
                        "ProductID": product_id,
                        "Name": name,
                        "Category": category,
                        "Price": price,
                        "Quantity": quantity
                    })
                add_products_in_bulk(inventory, products)

            elif choice == "9":
                product_ids = []
                while True:
                    product_id = input("Enter Product ID to delete (or 'done' to finish): ")
                    if product_id.lower() == 'done':
                        break
                    product_ids.append(product_id)
                delete_products_in_bulk(inventory, product_ids)

            elif choice == "10":
                print("Exiting the system. Goodbye!")
                break

            else:
                print("Invalid choice. Please try again.")

        except Exception as e:
            print(f"Error: {e}")

# Run the program
if __name__ == "__main__":
    main()
```

# Question 7 (20 points)

Create a well desined Python program for a flight booking system. Offer core functions for adding new flights, finding trips, reserving and releasing seating space and creating performance reports. Ensure the system manages edge cases like invalid input, overbooking, and cancellations. Finally, extend the booking system by including advanced features and maintain about 50 flight in the system.

**Prompt**

Give me a Python program that manages a flight booking system. The program does the following:

1. Loads and saves flight data from a CSV file.
2. Adds new flights, making sure there are no duplicates or missing information.
3. Searches for flights by destination.
4. Reserves and releases seats for passengers.
5. Updates flight statuses (Scheduled, Delayed, Canceled).
6. Generates a report on flight occupancy and status.
7. Initializes 50 sample flights if the file is empty.

```python
import csv
import os

# Define the CSV file path
FLIGHT_FILE = "flights.csv"

# Define the structure of the flight CSV file
CSV_HEADER = ["FlightNumber", "Destination", "TotalSeats", "AvailableSeats", "Status"]

# Function to load flights from CSV file
def load_flights():
    """
    Load flight data from the CSV file.
    Handles file not found and corrupted CSV files.
    """
    flights = []
    try:
        with open(FLIGHT_FILE, mode="r", newline="") as file:
            reader = csv.DictReader(file)
            # Edge Case 1: Corrupted CSV file (invalid headers)
            if reader.fieldnames != CSV_HEADER:
                raise ValueError("Corrupted CSV file: Invalid headers.")
            for row in reader:
                try:
                    row["TotalSeats"] = int(row["TotalSeats"])
                    row["AvailableSeats"] = int(row["AvailableSeats"])
                except ValueError:
                    raise ValueError("Corrupted CSV file: Invalid data types.")
                flights.append(row)
    except FileNotFoundError:
        # If the file doesn't exist, create it with headers
        with open(FLIGHT_FILE, mode="w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=CSV_HEADER)
            writer.writeheader()
    return flights

# Function to save flights to CSV file
def save_flights(flights):
    """
    Save flight data to the CSV file.
    """
    with open(FLIGHT_FILE, mode="w", newline="") as file:
        writer = csv.DictWriter(file, fieldnames=CSV_HEADER)
        writer.writeheader()
        writer.writerows(flights)

# Function to add a new flight
def add_flight(flights, flight_number, destination, total_seats):
    """
    Add a new flight to the system.
    Handles duplicate flight numbers and invalid inputs.
    """
    # Edge Case 2: Adding a flight that already exists
    for flight in flights:
        if flight["FlightNumber"] == flight_number:
            raise ValueError("Flight number already exists.")

    # Edge Case 3: Attempting to add a flight with invalid fields
    if not all([flight_number, destination, total_seats]):
        raise ValueError("Missing fields. All fields are required.")

    # Add the new flight
    flights.append({
        "FlightNumber": flight_number,
        "Destination": destination,
        "TotalSeats": int(total_seats),
        "AvailableSeats": int(total_seats),
        "Status": "Scheduled"
    })
    save_flights(flights)
    print(f"Flight {flight_number} to {destination} added successfully.")

# Function to find flights by destination
def find_flights(flights, destination):
    """
    Find flights by destination.
    Handles invalid input and returns matching flights.
    """
    if not destination:
        raise ValueError("Invalid destination.")

    found_flights = [flight for flight in flights if flight["Destination"] == destination]
    if not found_flights:
        return "No flights found for the given destination."
    return found_flights

# Function to reserve a seat
def reserve_seat(flights, flight_number, passenger_name):
    """
    Reserve a seat for a passenger.
    Handles edge cases like flight cancellation and overbooking.
    """
    for flight in flights:
        if flight["FlightNumber"] == flight_number:
            if flight["Status"] == "Canceled":
                return f"Cannot reserve seat on canceled flight {flight_number}."
            if flight["AvailableSeats"] > 0:
                flight["AvailableSeats"] -= 1
                save_flights(flights)
                return f"Seat reserved for {passenger_name} on Flight {flight_number} to {flight['Destination']}."
            else:
                return f"Flight {flight_number} is fully booked."
    return "Flight not found."

# Function to release a seat
def release_seat(flights, flight_number, passenger_name):
    """
    Release a reserved seat and update available seats.
    Handles edge case of non-existing reservations.
    """
    for flight in flights:
        if flight["FlightNumber"] == flight_number:
            flight["AvailableSeats"] += 1
            save_flights(flights)
            return f"Seat released for {passenger_name} on Flight {flight_number}."
    return "Flight not found."

# Function to update flight status (e.g., canceled, delayed)
def update_status(flights, flight_number, status):
    """
    Update the status of a flight.
    Handles invalid statuses and ensures correct updates.
    """
    valid_statuses = ["Scheduled", "Delayed", "Canceled"]
    if status not in valid_statuses:
        raise ValueError("Invalid status.")

    for flight in flights:
        if flight["FlightNumber"] == flight_number:
            flight["Status"] = status
            save_flights(flights)
            return f"Flight {flight_number} status updated to {status}."
    return "Flight not found."

# Function to generate a performance report
def generate_report(flights):
    """
    Generate a report on flight occupancy and status.
    Handles empty flights list and invalid data.
    """
    if not flights:
        raise ValueError("No flights available to generate a report.")

    report = []
    for flight in flights:
        occupancy_rate = (flight["TotalSeats"] - flight["AvailableSeats"]) / flight["TotalSeats"] * 100
        report.append(f"Flight {flight['FlightNumber']} to {flight['Destination']}: {occupancy_rate:.2f}% occupied. Status: {flight['Status']}.")
    return report

# Function to initialize flights with 50 sample entries
def initialize_flights():
    """
    Initialize the system with 50 flights.
    """
    flights = []
    destinations = ["New York", "Los Angeles", "Chicago", "San Francisco", "Miami"]
    for i in range(1, 51):
        flight_number = f"F{i:03}"
        destination = destinations[i % len(destinations)]
        total_seats = 100 + (i % 10) * 5
        flights.append({
            "FlightNumber": flight_number,
            "Destination": destination,
            "TotalSeats": total_seats,
            "AvailableSeats": total_seats,
            "Status": "Scheduled"
        })
    save_flights(flights)
    print("System initialized with 50 flights.")

# Main function to manage the flight booking system
def main():
    # Initialize flights with 50 flights if the file is empty
    if not os.path.exists(FLIGHT_FILE) or os.path.getsize(FLIGHT_FILE) == 0:
        initialize_flights()

    flights = load_flights()

    while True:
        print("\n--- Flight Booking System ---")
        print("1. Add Flight")
        print("2. Find Flights")
        print("3. Reserve Seat")
        print("4. Release Seat")
        print("5. Update Flight Status")
        print("6. Generate Report")
        print("7. Exit")

        choice = input("Enter your choice: ")

        try:
            if choice == "1":
                flight_number = input("Enter Flight Number: ")
                destination = input("Enter Destination: ")
                total_seats = input("Enter Total Seats: ")
                add_flight(flights, flight_number, destination, total_seats)

            elif choice == "2":
                destination = input("Enter Destination to find flights: ")
                found_flights = find_flights(flights, destination)
                if isinstance(found_flights, str):
                    print(found_flights)
                else:
                    for flight in found_flights:
                        print(f"Flight Number: {flight['FlightNumber']}, Destination: {flight['Destination']}, Total Seats: {flight['TotalSeats']}, Available Seats: {flight['AvailableSeats']}, Status: {flight['Status']}")

            elif choice == "3":
                flight_number = input("Enter Flight Number: ")
                passenger_name = input("Enter Passenger Name: ")
                print(reserve_seat(flights, flight_number, passenger_name))

            elif choice == "4":
                flight_number = input("Enter Flight Number: ")
                passenger_name = input("Enter Passenger Name: ")
                print(release_seat(flights, flight_number, passenger_name))

            elif choice == "5":
                flight_number = input("Enter Flight Number: ")
                status = input("Enter New Status (Scheduled, Delayed, Canceled): ")
                print(update_status(flights, flight_number, status))

            elif choice == "6":
                report = generate_report(flights)
                for line in report:
                    print(line)

            elif choice == "7":
                print("Exiting the system. Goodbye!")
                break

            else:
                print("Invalid choice. Please try again.")

        except Exception as e:
            print(f"Error: {e}")

# Run the program
if __name__ == "__main__":
    main()
```
